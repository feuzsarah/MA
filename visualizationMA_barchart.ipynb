```python
#simple barchart visualization

import matplotlib.pyplot as plt

#define the entries
method = ['m1', 'm2','m3','m4','m5','Benchmark']
values = [4.634, 4.438, 2.882, 2.894,4.243,4.612]
color = ['red', 'yellow', 'green', 'orange','gray' ,'blue']

#plot figure
plt.figure(figsize=(10,4))
plt.bar(method,values, width = 0.4, color = color)

#define layout of figure and save it
plt.xlabel('Methods')
plt.ylabel('$\sum$')
plt.title('Title')
plt.savefig('barchart.pdf')
```
